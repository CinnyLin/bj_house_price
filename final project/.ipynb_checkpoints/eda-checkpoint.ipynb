```python
import pandas as pd
import numpy as np

import statsmodels.api as sm
from sklearn import model_selection
import math

import matplotlib.pyplot as plt
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
#import descartes
#import folium
```

```python
df = pd.read_csv('1.BeijingHouse.csv')
df.head(3)
```

```python
df.columns
```

- `url`: the url which fetches the data
- `id`: the id of transaction
- `Lng`: latitude
- `Lat`：longitude
- `Cid`: community id
- `tradeTime`: the time of transaction
- `DOM`: active days on market.Know more in https://en.wikipedia.org/wiki/Days_on_market
- `followers`: the number of people follow the transaction.
- `totalPrice`: the total price
- `price`: the average price by square
- `square`: the square of house
- `livingRoom`: the number of living room
- `drawingRoom`: the number of drawing room
- `kitchen`: the number of kitchen
- `bathroom`: the number of bathroom
- `buildingType`: including tower(`1`) , bungalow(`2`)，combination of plate and tower(`3`), plate(`4`).
- `constructionTime`: the time of construction
- `renovationCondition`: including other(`1`), rough(`2`), Simplicity(`3`), hardcover(`4`)
- `buildingStructure`: including unknow(`1`), mixed(`2`), brick and wood(`3`), brick and concrete(`4`),steel(`5`) and steel-concrete composite (`6`).
- `ladderRatio`: the proportion between number of residents on the same floor and number of elevator of ladder. It describes how many ladders a resident have on average.
- `Elevator`： have (`1`) or not have elevator(`0`)
- `fiveYearsProperty`: if the owner have the property for less than 5 years.
- `subway` ：near subway station (`1`) or not (`0`)

```python
crs = {'init': 'epsg:4326'}
geometry = [Point(xy) for xy in zip(df['Lng'], df['Lat'])]
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
street_map.plot(ax=ax, alpha=0.4, color='grey')
geo_df[geo_df['fiveYearsProperty']==0].plot(ax=ax, color='red', label='0')
geo_df[geo_df['fiveYearsProperty']==1].plot(ax=ax, color='blue', label='1')
plt.legend()
plt.title('Whether owner have the property for more than five years')
plt.show()
```

```python
# ----- READ IN DATA -----
df_train, df_test = model_selection.train_test_split(df, test_size=0.3)
```

```python
# specify which column contains outcome variable y. Needs to be coded as binary True/False or 0/1
# dropped columns: 'tradeTime', 'constructionTime', 'buildingType'

x_col_name_num = ['Lng', 'Lat', 'Cid', 'DOM', 'followers',
                  'totalPrice', 'square', 'livingRoom', 'drawingRoom', 'kitchen',
                  'bathRoom', 'renovationCondition', 'buildingStructure', 'ladderRatio',
                  'subway', 'elevator', 'fiveYearsProperty']
y_col_name_num = ['price']
```

```python
# --- fit linear probability model
lpm_model = sm.OLS(endog=df_train[y_col_name_num], exog=df_train[x_col_name_num])
lpm_results = lpm_model.fit()
print("\nRESULTS OF LOGIT MODEL FITTING\n")
print(lpm_results.summary())
```

```python
# specify which column contains outcome variable y. Needs to be coded as binary True/False or 0/1
# dropped columns: 'tradeTime', 'constructionTime', 'buildingType'
# dropped columns: 'Lng', 'Lat', 'Cid', 'DOM', 'followers', 'price'
#                  'totalPrice', 'square', 'livingRoom', 'drawingRoom', 'kitchen',
#                  'bathRoom', 'renovationCondition', 'buildingStructure', 'ladderRatio'

x_col_name_bi = ['subway', 'elevator', 'fiveYearsProperty']
y_col_name_bi = ['avg_price']
```

```python
# --- fit linear probability model
lpm_model = sm.OLS(endog=df_train[y_col_name_bi], exog=df_train[x_col_name_bi])
lpm_results = lpm_model.fit()
print("\nRESULTS OF LOGIT MODEL FITTING\n")
print(lpm_results.summary())
```

```python
# --- fit logistic regression
logit_model = sm.Logit(endog=df_train[y_col_name_bi], exog=df_train[x_col_name_bi])
logit_results = logit_model.fit()
print("\nRESULTS OF LOGIT MODEL FITTING\n")
print(logit_results.summary())
```

```python
# ----- EVALUATE MODEL PERFORMANCE -----
# --- evaluate model predictions on test set graphically
df_test['prob_price_lpm'] = lpm_results.predict(df_test[x_col_name_bi])
df_test['prob_price_logit'] = logit_results.predict(df_test[x_col_name_bi])
```
